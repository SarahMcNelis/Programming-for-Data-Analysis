## Supply and Demand

There is a clear link here between infection, mortality and vaccination. However, another factor that have become evident is supply and demand. Some countries with a higher population have a higher case and death rate than those with lower population. But how are vaccines being allocated?


### Allocation of vaccines

The European Commission presented their vaccination strategy found [here](https://ec.europa.eu/info/live-work-travel-eu/coronavirus-response/public-health/eu-vaccines-strategy_en#:~:text=On%2019%20January%202021%2C%20the,Member%20State%20should%20get%20vaccinated.) on June 17th 2020. This strategy included development, manufacturing and deployment of covid-19 vaccines within the EU. The EU commission invested 2.7 billion euro through their emergency support scheme in order to secure covid-19 vaccine contracts on behalf of EU member states. An important element of this strategy was the allocation of vaccines. This [allocation](https://www.euronews.com/2021/04/02/how-many-vaccine-doses-have-arrived-in-eu-countries-and-how-do-they-get-there) was based per population of each country. Therefore, the idea was that each country would recieve enough vaccine to vaccinate the same percentage of their population on the same timeframe. The hope was that most EU member states would have recieved enough vaccine to have at least 10% of their population fully vaccinated by March 2021. This would suggest a [uniform distribution](https://en.wikipedia.org/wiki/Continuous_uniform_distribution) plan for the allocation of vaccines. In other words, each country would have equal opportunity of reaching this target. 

<br>

#### Fully vaccinated on March 31st 2021

<img src="img/fully_vaccinated_march2021.png" width="900"/>

- Looking at the end of March 2021, we can see Malta is the only country to have fully met the desired outcome. The majority of the EU memeber states had around half of the desired percentage fully vaccinated. The question remains why was the proposed target not met?

- Upon further research I discovered it the responsibility of each member state to decide if they wanted to buy the full amount of allocated doses. If a country did not secure the full amount, then this surplus allocation would be redistributed among the other member states. This would explain why some countries had a higher vaccination percentage than others. 

- Therefore, it is fair to say that even though each EU member state was given equal opportunity, this desired uniform distribution was not achieved.

- Set the [index](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.index.html) to begin at 1 instead of the inital default index value of 0. 
- Check that the data was loaded into the dataframe by calling df. 

```python
# Create data.
country = {'country':['Austria', 'Belgium', 'Bulgaria', 'Croatia', 'Cyprus', 'Czech Republic', 'Denmark',
           'Estonia', 'Finland', 'France', 'Germany', 'Greece', 'Hungary', 'Ireland', 'Italy', 'Latvia',
           'Lithuania', 'Luxembourg', 'Malta', 'Netherlands', 'Poland', 'Portugal', 'Romania', 'Slovakia',
           'Slovenia', 'Spain', 'Sweden']}

# Load data to df.
df = pd.DataFrame(country)
# Name column.
df.columns = ['country']
# Set index to start at 1. 
df.index = np.arange(1, len(df)+1)
# Display df to check it worked. 
df
```

<br>

### **Step 2:**
- Set the [seed](https://www.geeksforgeeks.org/random-seed-in-python/) value in order to designate the same start point when generating the random data. This will deliver the same random data each time the code is executed. 
- Create more data for a new dataframe. 
- Using the [random randint](https://numpy.org/doc/stable/reference/random/generated/numpy.random.randint.html) function to return 27 random integers within the set range. 
- Using [random uniform](https://numpy.org/doc/stable/reference/random/generated/numpy.random.uniform.html) to return random floats in given the range in uniform distribution.
- Calling the [numpy around](https://numpy.org/doc/stable/reference/generated/numpy.around.html) funtion to round the floats in the arrays to 2 decimal places. 
- Next I pass the created data into df2, setting the column names and index again. 
- Check it successfully loaded. 

```python
# Setting inital start point to generate same random numbers. 
np.random.seed(1000)

# Creating random integers and floats. 
data = {'popultion': np.random.randint(470000, 83000000, 27), 
        'reported_cases': np.random.randint(40624, 400000, 27),
        'reported_deaths': np.random.randint(471, 135000, 27),
        'fully_vaccinated': np.around(np.random.uniform(26.67, 88.81, 27), 2),
        'partially_vaccinated': np.around(np.random.uniform(0.21, 6.28, 27), 2)}
    
# Inputting data into df2. 
df2 = pd.DataFrame(data)
# Setting column names. 
df2.columns = ['population', 'reported_cases', 'reported_deaths', 'fully_vaccinated',
              'partially_vaccinated']
# Setting same index as df.
df2.index = np.arange(1, len(df2)+1)
df2
```

<br>

### **Step 3:**
- Using the [pandas join](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.join.html) function here to join the two dataframes together. This will combine the countries with the random data I created above to give us a clear view of what the covid-19 figures for each country may look like.

```python
# Joining df2 to df.
df = df.join(df2)
df
```

<br>

### **Step 4:**
- Creating new data using the calcuations created in df2. 
- Using [numpy add](https://numpy.org/doc/stable/reference/generated/numpy.add.html) to add the vaccinated and partially vaccinated together.
- Then using [numpy subtract](https://numpy.org/doc/stable/reference/generated/numpy.subtract.html) function to subtract the new values from 100. 
- Loading this into another df.
- Joining this df to the origional df. 

```python
# using np function to subtract one array from another. 
unvaccinated = np.add(df2['fully_vaccinated'], df2['partially_vaccinated'])
unvaccinated = np.subtract(100, unvaccinated)

# Load df3. 
df3 = pd.DataFrame(unvaccinated)
# Setting column name. 
df3.columns = ['unvaccinated']
# Set same index again.
df3.index = np.arange(1, len(df3)+1)
```

```python
# Check it worked. 
df3
```

```python
# Join to origional df. 
df = df.join(df3)
df
```

<br>

### **Step 5:**
- Creating a column showing the cumulative percentage of population with confirmed covid-19 cases.
- Using [numpy divide](https://numpy.org/doc/stable/reference/generated/numpy.divide.html) and [numpy multiply](https://numpy.org/doc/stable/reference/generated/numpy.multiply.html) to create the percentage. 
- Using around funciton again to round to 2 decimal places. 
- Load into df and check it worked. 
- Join to origional df. 

```python
# Get percentage of pop with cases. 
cases_percentage = np.divide( df2['reported_cases'],df2['population'])
cases_percentage = np.around((np.multiply(cases_percentage, 100)), 2)

# Load df4. 
df4 = pd.DataFrame(cases_percentage)
# Setting column name. 
df4.columns = ['cases_percentage']
# Set same index again.
df4.index = np.arange(1, len(df4)+1)
df4
```

```python
# Join to origional df and check. 
df = df.join(df4)
df
```

<br>

### **Step 6:**
- Create the cumulative percentage of population with confirmed covid-19 related deaths. 
- Using same method as step 5; numpy divide, multiply and around. 
- Load the data and check it was successful. 
- Join to origional df. 

```python
# Create more data from generated data. 
death_percentage = np.divide( df2['reported_deaths'],df2['population'])
death_percentage = np.around((np.multiply(death_percentage, 100)), 2)

# Load df5. 
df5 = pd.DataFrame(death_percentage)
# Setting column name. 
df5.columns = ['death_percentage']
# Set same index again.
df5.index = np.arange(1, len(df4)+1)
df5
```

```python
# Join to origional df. 
df = df.join(df5)
```
